It is also quite interesting, because, as we can see, at the end of the given period, the amount of terrorit attacks has peaked. This could mean that the goverment of India has invested more money in their citizens safety.

Targets, that are being attacked more often, are Private Citizens & Property, Police and Government. For each of them the most frequently used weapon are firearms. Also it seems that terrorists in India aren't interested in foreigners because there aren't many cases of airports and tourists being attacked.
